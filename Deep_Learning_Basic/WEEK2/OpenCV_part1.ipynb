## Open CV

```python
import cv2
cv2.__version__
```

```python
img_path = 'images/image_1.png'
img_read = cv2.imread(img_path)
print('Shape of the image:', img_read.shape)
```

```python
from matplotlib import pyplot as plt
plt.imshow(img_read)
plt.show()
```

```python
cv2.imwrite('my_save_pic.png',img_read)
```

```python
# crop image using numpy slice
plt.imshow(img_read[30:200, 100:280])
plt.show()
```

```python
# again crop image
plt.imshow(img_read[225:325, 50:325])
plt.show()
```
